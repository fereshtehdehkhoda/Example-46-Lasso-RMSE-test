```python
#روی داده‌های آموزش hw12_train.csv یک مدل Lasso Regression بسازید. توان‌های 2، 3 ،4 و 5 متغیر x را بعنوان ویژگی‌های جدید به ماتریس ویژگی‌ها اضافه کنید.

#به منظور پیدا کردن بهترین عدد برای فراپارامتر مدل، الگوریتم k-Fold Cross Validation را بکار بگیرید. تعداد پوشه‌ها را 10 در نظر بگیرید. مقادیر کاندیدا برای فراپارامتر به شکل زیر تعریف شود:

#10 ** np.linspace(3, -3, 100)

#برای بهترین مدل، مقدار شاخص RMSE روی داده‌های آزمایش hw12_test.csv چقدر است؟
```

```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('hw12_train.csv')
```

```python
data['x^2']=data['x'] ** 2
data['x^3']=data['x'] ** 3
data['x^4']=data['x'] ** 4
data['x^5']=data['x'] ** 5
```

```python
features=['x','x^2','x^3','x^4','x^5']
X_train=data[features]

target='y'
y_train=data[target]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
```

```python
alpha_gride=np.linspace(3,-3,100) ** 10
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Lasso
cv_errors=np.zeros(shape=len(alpha_gride))

for i in range(len(alpha_gride)):
    lasso=Lasso(alpha=alpha_gride[i])
    scores=cross_val_score(estimator=lasso,
                          X=X_train_scaled,
                          y=y_train,
                          scoring='neg_root_mean_squared_error',
                          cv=10,n_jobs=-1)
    cv_errors[i]=scores.mean()

best_alpha=alpha_gride[np.argmax(cv_errors)]
best_alpha
```

```python
from sklearn.linear_model import Lasso
model_1=Lasso(alpha=best_alpha)
model_1.fit(X_train_scaled,y_train)
```

```python
dtest=pd.read_csv('hw12_test.csv')
```

```python
dtest['x^2']=dtest['x'] ** 2
dtest['x^3']=dtest['x'] ** 3
dtest['x^4']=dtest['x'] ** 4
dtest['x^5']=dtest['x'] ** 5
```

```python
features=['x','x^2','x^3','x^4','x^5']
X_test=dtest[features]

target='y'
y_test=dtest[target]
```

```python
X_test_scaled=scaler.transform(X_test)
```

```python
y_test_pred_1=model_1.predict(X_test_scaled)

from sklearn.metrics import root_mean_squared_error,mean_absolute_percentage_error
print("RMSE:",round(root_mean_squared_error(y_test_pred_1,y_test),3))
print("MAPE:",round(mean_absolute_percentage_error(y_test_pred_1,y_test)* 100,2))
```
